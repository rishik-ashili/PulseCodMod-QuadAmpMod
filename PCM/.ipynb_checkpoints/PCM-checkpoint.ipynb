```python
import sounddevice as sd
from scipy.io.wavfile import write
import numpy as np

def record_voice(filename, duration, fs=44100):
    """
    Records voice input and saves it as a .wav file.

    Parameters:
    - filename: Name of the output .wav file (string).
    - duration: Recording duration in seconds (integer).
    - fs: Sampling rate (default: 44100 Hz).
    
    """
    print("Recording...")
    # Record the audio with given duration and sample rate
    audio = sd.rec(int(duration * fs), samplerate=fs, channels=1)
    sd.wait()  # Wait until the recording is done
    print("Recording finished.")
    
    # Scale to int16 format and save as .wav
    audio_int16 = np.int16(audio * 32767)
    write(filename, fs, audio_int16)
    print(f"Audio saved as {filename}")

# Example usage
# record_voice('output.wav', duration=5, fs=44100)  # Records for 5 seconds
```

```python
record_voice("adarshhhhhhhh",5)
```

```python
import sounddevice as sd
import soundfile as sf

def play_audio(filename):
    """
    Plays a .wav audio file.

    Parameters:
    - filename: Path to the .wav file (string).
    """
    # Read the audio file
    audio_data, fs = sf.read(filename)
    
    print(f"Playing {filename}...")
    # Play the audio
    sd.play(audio_data, samplerate=fs)
    sd.wait()  # Wait until the audio playback is finished
    print("Playback finished.")

# Example usage
# play_audio('output.wav')  # Plays the 'output.wav' file
```

```python
play_audio("adarshhhhhhhh")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf

def plot_waveform(filename):
    """
    Plots the waveform of a .wav audio file.

    Parameters:
    - filename: Path to the .wav file (string).
    """
    # Read the audio file
    audio_data, fs = sf.read(filename)
    
    # Create time axis in seconds
    time_axis = np.linspace(0, len(audio_data) / fs, num=len(audio_data))
    
    # Plot the waveform
    plt.figure(figsize=(15, 5))
    plt.plot(time_axis, audio_data, color='blue')
    plt.title(f"Waveform of {filename}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
# plot_waveform('output.wav')  # Plots the waveform of 'output.wav'
```

```python
plot_waveform("adarshhhhhhhh")
```

```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt

def perform_pcm(filename):
    try:
        # Read the audio file
        audio_data, sample_rate = sf.read(filename)
        print(f"Audio file '{filename}' loaded successfully. Sample rate: {sample_rate}", flush=True)

        # Normalize audio data
        min_value = np.min(audio_data)
        max_value = np.max(audio_data)
        normalized_audio = (audio_data - min_value) / (max_value - min_value)

        # Set PCM parameters
        bit_depth = 4
        quantization_levels = 2 ** bit_depth

        # Perform quantization
        quantized_audio = np.round(normalized_audio * (quantization_levels - 1)) / (quantization_levels - 1)

        # Save the quantized PCM output as a WAV file
        output_filename = 'quantized_pcm_output.wav'
        sf.write(output_filename, quantized_audio, sample_rate)
        print(f"Quantized PCM output saved as '{output_filename}'.", flush=True)

        # Create binary PCM representations
        binary_pcm = np.array([np.binary_repr(int(s * (quantization_levels - 1)), width=bit_depth) for s in quantized_audio])

        # Plotting the quantized audio and binary PCM values
        plot_waveform(filename)
        plt.figure(figsize=(15, 10))
        # Plot Binary Representation (Use integers instead of strings for plotting)
        plt.subplot(2, 1, 2)
        int_pcm_values = [int(b, 2) for b in binary_pcm]  # Convert binary strings to integers
        plt.plot(int_pcm_values, label='PCM Audio Signal (Integer Representation)', color='blue')
        plt.title('PCM Audio Signal (Binary Representation as Integer)')
        plt.xlabel('Sample Index')
        plt.ylabel(f'PCM Integer Values (0 to {quantization_levels-1})')
        plt.grid(True)
        plt.legend()

        # Display the plots
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}", flush=True)

# Example usage

```

```python
# pcm_on_audio_file("adarshhhhhhhh")
perform_pcm("adarshhhhhhhh")
```

```python
play_audio("quantized_pcm_output.wav")
```

```python
import numpy as np
import soundfile as sf

def demodulate_pcm(filename, output_filename='demodulated_output.wav', bit_depth=4):
    # Load the PCM-modulated audio file
    pcm_audio, sample_rate = sf.read(filename)

    # Calculate the number of quantization levels
    quantization_levels = 2 ** bit_depth

    # Reverse the quantization process: Map discrete values back to their range
    demodulated_audio = pcm_audio * (quantization_levels - 1)  # Denormalize to discrete levels

    # Now, scale it back to a normalized range (0 to 1), matching the modulation process
    demodulated_audio /= (quantization_levels - 1)

    # Save the demodulated audio as a WAV file
    sf.write(output_filename, demodulated_audio, sample_rate)
    print(f"Demodulated audio saved as '{output_filename}'.")
```

```python
demodulate_pcm("quantized_pcm_output.wav")
```
